```python
from IPython.display import display
from PIL import Image
from google.colab import files  # For uploading images in Colab

# Upload images manually if needed
# files.upload()

# Load images
images = {
    "treasureisland": Image.open("Day3_TreasureIslandImage.jpg").resize((200, 200)),
    "crossroads": Image.open("Day3_CrossRoads.jpg").resize((200, 200)),
    "lake": Image.open("Day3_lake.jpg").resize((200, 200)),
    "boat":Image.open("Day3_boat.jpg").resize((200, 200)),
    "house": Image.open("Day3_house.jpg").resize((200, 200)),
    "swim": Image.open("Day3_swim.jpg").resize((200, 200)),
    "hole": Image.open("Day3_hole.jpg").resize((200, 200)),
    "gameover":Image.open("Day3_GameOver.jpg").resize((200, 200)),
    "fire":Image.open("Day3_fire.jpg").resize((200,200)),
    "treasure":Image.open("Day3_treasure.jpg").resize((200,200)),
    "beast":Image.open("Day3_beast.jpg").resize((200,200))
}
# Start Game
print("Welcome to Treasure Island")
print("Your mission is to find the treasure.")
display(images["treasureisland"])
print("\n\n\n")
print("You are at a crossroad.")
display(images["crossroads"])
print("\n\n\n")
cross_choice = input("Where do you want to go? \nType Left or Right: ").strip().lower()
if cross_choice == "left":
  print("You have reached the lake.")
  display(images["lake"])
  lake_choice = input("\nThere is an island in the middle of the lake.\nType 'wait' for a boat or 'swim' to swim across: ").strip().lower()
  print("\n\n\n")
  if lake_choice == "wait":
    display(images["boat"])
    print("You have arrived at the island unharmed.")
    print("\n\n\n")
    display(images["house"])
    door_choice=input("There is a house with 3 doors: \nRed, Yellow, Blue.Which color do you choose?").strip().lower()
    print("\n\n\n")
    if door_choice == "red":
      display(images["fire"])
      print("🔥 FIRE!!!! You have entered a room full of fire. GAME OVER.")
      print("\n\n\n")
    elif door_choice == "yellow":
      display(images["treasure"])
      print("🎉 HURRAY!!! You found the treasure. YOU WON! 🎊")
      print("\n\n\n")
    elif door_choice == "blue":
      display(images["beast"])
      print("🐺 BAD!!! You entered a room of beasts. GAME OVER.")
      print("\n\n\n")
    else:
      display(images["gameover"])
      print("❌ Wrong choice. GAME OVER.")
      print("\n\n\n")
  elif lake_choice == "swim":
    display(images["swim"])
    print("🐊 OHHHH!!! You are attacked by an alligator. GAME OVER.")
    print("\n\n\n")
  else:
    display(images["gameover"])
    print("❌ Invalid choice. GAME OVER.")
    print("\n\n\n")
elif cross_choice == "right":
  display(images["hole"])
  print("💀 OOPS!!! You fell into a hole. GAME OVER.")
  print("\n\n\n")
else:
  display(images["gameover"])
  print("❌ Invalid choice. GAME OVER.")
```
