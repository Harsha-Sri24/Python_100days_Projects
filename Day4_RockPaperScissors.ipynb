```python
from IPython.display import display
from PIL import Image
import random
from google.colab import files
images={
    "game": Image.open("Day4_game.jpg").resize((200,200)),
    0: Image.open("Day4_rock.jpg").resize((200,200)),
    1: Image.open("Day4_paper.jpg").resize((200,200)),
    2: Image.open("Day4_scissors.jpg").resize((200,200))

}
print("Welcome to rock paper scissors")
display(images["game"])
user_choice=int(input("Type 0 for rock \n 1 for paper \n 2 for scissors"))
computer_choice=random.randint(0,2)
if user_choice>2:
  print("Invalid number!!")
else:
  print("user chose")
  display(images[user_choice])
  print("\n\n\n")
  print("computer chose")
  display(images[computer_choice])
  if user_choice==0 and computer_choice==2:
    print("You win")
  elif computer_choice==0 and user_choice==2:
    print("You lose")
  elif computer_choice>user_choice:
    print("You lose")
  elif user_choice>computer_choice:
    print("you win")
  else:
    print("It's a draw")

```
